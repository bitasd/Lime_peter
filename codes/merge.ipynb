```python
import pandas, json, os, sys
```

```python
def main(tgp):
    
    DATA_PATH=f'C:\\Users\\bitas\\folders\\Research\\lime\\peter\\Data\\detour\\{tgp}\\json_files\\'
    result = []
    fps = [os.path.join(DATA_PATH, f) for f in os.listdir(DATA_PATH)]
    for fp in fps:
        print(fp)
        with open(fp) as f:
            print(f)
            data = json.load(f)
            features_lst = data['features']
            result = result + features_lst
    print(len(result))
    linesCollection = dict(type="FeatureCollection",features=result)
    json_temp = json.dumps(linesCollection)

    

    f = open(f"C:\\Users\\bitas\\folders\\Research\\lime\\peter\\Data\\detour\\{tgp}\\{tgp}.json","w")
    f.write(json_temp)
    f.close()
```
