```python
def reproject_(line, init_proj, dest_proj):
    tfm = partial(pyproj.transform, 
                  pyproj.Proj(init=init_proj), 
                  pyproj.Proj(init=dest_proj))
    reprojected = ops.transform(tfm, line)
    return reprojected
```

```python
class Agent(object):
    
    def __init__(self, data):
        
        self.data = data
        self.trips_legs = []
        
    def processBatch(self, idx_lst):
        for idx in idx_lst:
            self.process(idx)
            
        
    def process(self, idx):
        
        trip_id = self.data[idx]['properties']['trip_id']
        coords = self.data[idx]['geometry']['coordinates']
        line_ = LineString(coords)
        line_re = reproject_(line_, "epsg:4326","epsg:26986")
        coords_re = list(line_re.coords)
        legs_coords = [[coords_re[i], coords_re[i+1]] for i in range(len(coords_re)-1) ]
        length = [LineString(i).length for i in legs_coords]
        elev_diff = [(i[1][-1]-i[0][-1]) for i in legs_coords]
        legs_dict = {'trip_id':trip_id, 'legs':legs_coords, 'leg_len':length, 'elev_diff':elev_diff}
#         print(legs_dict)

    
        self.trips_legs.append(legs_dict)
#         print(self.trips_legs)
    
    def exportTOcsv(self, counter):
        
        processed_batch = self.trips_legs
        df = pandas.DataFrame()
        for trip in processed_batch:

            df = df.append(pandas.DataFrame(trip))
            print(len(df))
            df.to_csv(f"C:\\Users\\bitas\\folders\\Research\\lime\\Data\\scooter\\foot\\csv\\{counter}.csv")
```
